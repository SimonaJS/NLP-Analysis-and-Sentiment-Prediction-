### Introduction

In the competitive hospitality industry, customer satisfaction is a key factor influencing business success. For hotels, understanding the sentiments and needs of guests can help enhance service quality, improve customer experience, and drive positive reviews and repeat bookings. However, traditional feedback methods often overlook the rich insights embedded within customer reviews. Sentiment analysis using natural language processing (NLP) provides an opportunity to analyze large volumes of unstructured review data, transforming it into actionable insights.

This project aims to leverage machine learning to develop a predictive model that can analyze customer reviews and identify key factors contributing to positive and negative sentiments. By focusing on customer satisfaction indicators, the project offers a valuable tool for hotel managers to pinpoint strengths and weaknesses in their services, thus supporting data-driven decision-making.

### Objective 

Hotels and other accommodation providers rely on customer reviews to gain insights into guest experiences. However, manually analyzing thousands of reviews is impractical and may overlook critical patterns. Negative reviews can damage a hotel’s reputation, while positive reviews strengthen brand value. Hospitality Insights Group  seeks to enhance its understanding of customer satisfaction by identifying the key factors that influence guest ratings.

This project aims to analyze a comprehensive dataset from Hospitality Insights Group, containing hotel reviews, stay details (such as location and duration), and a target column (`reviewer_score`) indicating overall satisfaction (1 for positive, 0 for negative). By developing a machine learning model to `predict reviewer_score`, this project will reveal actionable insights into customer satisfaction drivers. The resulting model will empower Hospitality Insights Group to proactively address areas for improvement, tailor services to guest needs, and ultimately enhance the customer experience, fostering loyalty and competitive advantage.

### Dataset Description 

The dataset used in this project is a comprehensive collection of hotel reviews for Hospitality Insights Group and shared as an Excel file through BrainStation. It contains 515,738 instances and 29 features, offering an in-depth look at factors that may influence customer satisfaction and hotel ratings. The dataset is particularly relevant to the field of customer experience analysis, supporting both natural language processing (NLP) and classification tasks.

This dataset includes 11 binary features, 10 continuous features and 2 text-based features which provide free-text comments from guests regarding their experiences, allowing for sentiment analysis and NLP-driven insights. Each instance in the dataset represents a unique hotel stay, with detailed information on both quantitative and qualitative aspects of the guest experience. The combination of numerical, categorical, and textual data makes this dataset a rich resource for building machine learning models aimed at predicting customer satisfaction. Specifically, the inclusion of review content in text format introduces opportunities to explore sentiment analysis, assess word patterns, and identify key drivers of positive and negative sentiment.

### Sentiment Analysis 

Now we will run our sentiment analysis using `TextBlob`, which assigns a polarity score between -1 (very negative) and 1 (very positive), with a score of 0 indicating neutral sentiment.  

```python
from textblob import TextBlob

# Start timing
start_time = time.time()

# Making a copy of the original DataFrame for sentiment analysis
df_sentiment = df.copy()

# Calculating sentiment only if the review is not a placeholder
df_sentiment['positive_sentiment'] = df_sentiment['positive_review'].apply(lambda x: TextBlob(x).sentiment.polarity if x != 'No Positive' else None)
df_sentiment['negative_sentiment'] = df_sentiment['negative_review'].apply(lambda x: TextBlob(x).sentiment.polarity if x != 'No Negative' else None)

# Defining thresholds
positive_threshold = 0.3
negative_threshold = -0.3

# Ensuring thresholds only apply to non-null sentiment values
df_sentiment['positive_match'] = ((df_sentiment['reviewer_score'] == 1) & (df_sentiment['positive_sentiment'] >= positive_threshold) & (df_sentiment['positive_sentiment'].notnull()))
df_sentiment['negative_match'] = ((df_sentiment['reviewer_score'] == 0) & (df_sentiment['negative_sentiment'] <= negative_threshold) & (df_sentiment['negative_sentiment'].notnull()))

# Calculating the proportion of matches
positive_alignment_rate = df_sentiment['positive_match'].mean() * 100
negative_alignment_rate = df_sentiment['negative_match'].mean() * 100

# Calculating total running time
end_time = time.time()
total_time = end_time - start_time

# Printing results
print(f"Proportion of reviews with reviewer_score = 1 that meet positive sentiment threshold: {positive_alignment_rate:.2f}%")
print(f"Proportion of reviews with reviewer_score = 0 that meet negative sentiment threshold: {negative_alignment_rate:.2f}%\n")
print(f"Total running time: {total_time:.2f} seconds")
```

##### Summary of Sentiment Analysis Findings

The analysis reveals a notable misalignment between reviewer scores and sentiment polarity from review text:

1. ***Positive Reviews***: 41.04% of reviews with a positive reviewer score (1) contain positive sentiment, suggesting that nearly 59% of positively scored reviews include mixed or critical feedback. This misalignment implies that positive reviewer scores might still reflect varied customer experiences or nuanced opinions not fully captured by sentiment alone.
<br>

2. ***Negative Reviews***: Only 5.32% of reviews with a negative score (0) reflect negative sentiment in the text, indicating that most low scores are not strongly tied to negative language. This further suggests that factors other than text sentiment, such as location or stay length, may influence the overall reviewer score.

These findings highlight that while reviewer scores do not align strictly with sentiment, review text remains a valuable component in predicting customer satisfaction. Through TF-IDF and n-grams, the model can extract essential keywords and themes, offering predictive insights that enrich the score when combined with other features. This approach can capture broader aspects of the customer experience that might otherwise be overlooked.

### Statistical Redundancy Analysis

In this section, we will conduct a Statistical Redundancy Analysis to identify and address any redundant features that may affect model performance and interpretability. Redundant features—those that overlap significantly in the information they provide—can lead to model overfitting, increased complexity, and decreased generalization. To evaluate redundancy, we will apply three key techniques.

#### Cramer’s V: Analyzing associations between categorical variables

Cramér’s V is a measure of association used to assess the strength of the relationship between two categorical variables. It provides values between 0 (no association) and 1 (strong association), indicating how strongly categorical variables are linked. By examining pairs of categorical features, we can determine if any are highly associated, suggesting redundancy that may allow for feature reduction.

```python
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    contingency_table = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(contingency_table)
    n = contingency_table.sum().sum()
    r, k = contingency_table.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# Selecting only binary features for Cramér's V calculation
binary_features = ['leisure_trip', 'reviewer_nationality_australia', 'reviewer_nationality_ireland', 
                   'reviewer_nationality_uae', 'reviewer_nationality_uk', 'reviewer_nationality_usa', 
                   'hotel_address_amsterdam', 'hotel_address_barcelona', 'hotel_address_milan', 
                   'hotel_address_paris', 'hotel_address_vienna']

# Creating a DataFrame to store Cramér's V results
cramers_v_matrix = pd.DataFrame(index=binary_features, columns=binary_features)

# Calculating Cramér's V for each pair of binary features
for feature1 in binary_features:
    for feature2 in binary_features:
        if feature1 != feature2:
            cramers_v_matrix.loc[feature1, feature2] = cramers_v(df[feature1], df[feature2])
        else:
            cramers_v_matrix.loc[feature1, feature2] = np.nan  # Avoid self-comparison

# Displaying the Cramér's V matrix with formatted values
cramers_v_matrix = cramers_v_matrix.astype(float)
cramers_v_matrix.style.background_gradient(cmap='YlOrRd').format(precision=2)
```

### Custom Text Tokenization

In this section, we will prepare our text features (`positive_review` and `negative_review`) for modelling by applying custom text preprocessing steps. This preprocessing is essential for creating meaningful tokens that can be vectorized effectively, which will allow us to capture important patterns in the data for model training.

The custom tokenizer we define here will process the review text by:

- Splitting sentences into individual words for detailed analysis.
- Removing special characters and filtering out non-informative stopwords to focus on relevant content.
- Applying both stemming and lemmatization to reduce words to their core forms while retaining context.

To ensure the model can distinguish between positive and negative feedback, we add prefixes (`pos_` for positive reviews and `neg_` for negative reviews) to each token. This approach will help the model learn sentiment-based patterns within the text data, enhancing its ability to make accurate predictions.

```python
# Importing required libraries
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer

# Defining my tokenizer function
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def my_tokenizer(text, review_type):
    """
    Tokenizes the input text by:
    - Removing punctuation and lowering case,
    - Splitting into words,
    - Filtering out stopwords and non-alphabetic characters,
    - Stemming and lemmatizing each word,
    - Adding a prefix based on review type ('pos_' or 'neg_').
    """
    # Removing punctuation and lowering the text
    text = re.sub(r'[^\w\s]', '', text).lower() 
    
    # Splitting text into words
    tokens = text.split()  
     
    # Removing stopwords and non-alphabetic characters
    english_stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in english_stop_words and word.isalpha()] 
    
    # Stemming each word
    tokens = [stemmer.stem(word) for word in tokens]  
    
    # Lemmatizing each word
    tokens = [lemmatizer.lemmatize(word) for word in tokens] 
    
    # Prefixing words based on review type
    tokens = [f'{review_type}_{word}' for word in tokens]
    return tokens
```

#### Preprocessing: Vectorizing Text Features and Concatenating Processed Datasets

In this section, we transform our text data into a numerical format suitable for machine learning. Using `CountVectorizer`, we convert the `combined_review` text column into vectorized features, capping the number at 500 to balance model complexity and interpretability. After vectorization, we combine these text features with our standardized numerical and binary features, creating a fully prepared dataset ready for training, validation, and testing.

```python
import time
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import CountVectorizer

# Starting time
start_time = time.time()

# Scaling non-binary numerical features
scaler = StandardScaler()

# Defining the CountVectorizer with a custom tokenizer
text_transformer = CountVectorizer(
    tokenizer=lambda text: my_tokenizer(text, 'pos') + my_tokenizer(text, 'neg'),
    max_features=500, min_df=10, token_pattern=None 
)

# Fitting and transforming numerical features separately for train, validation, and test sets
X_train_numerical_scaled = scaler.fit_transform(X_train[numerical_features])
X_val_numerical_scaled = scaler.transform(X_val[numerical_features])
X_test_numerical_scaled = scaler.transform(X_test[numerical_features])

# Fitting and transforming text features
X_train_text_cv = text_transformer.fit_transform(X_train[text_features])
X_val_text_cv = text_transformer.transform(X_val[text_features])
X_test_text_cv = text_transformer.transform(X_test[text_features])

# Drop combined_review column from the original dataframes
X_train = X_train.drop(columns=['combined_review'])
X_val = X_val.drop(columns=['combined_review'])
X_test = X_test.drop(columns=['combined_review'])

# Concatenating numerical, text, and binary features for training, validation, and testing data
X_train_processed_cv = np.concatenate((X_train_numerical_scaled, X_train_text_cv.toarray(), X_train[binary_features]), axis=1)
X_val_processed_cv = np.concatenate((X_val_numerical_scaled, X_val_text_cv.toarray(), X_val[binary_features]), axis=1)
X_test_processed_cv = np.concatenate((X_test_numerical_scaled, X_test_text_cv.toarray(), X_test[binary_features]), axis=1)

# End time
end_time = time.time()

# Printing the total runtime for preprocessing
print(f"Processing runtime: {end_time - start_time:.2f} seconds")

# Printing shapes of train, validation, and test sets
print("\nShape of X_train_processed_cv:", X_train_processed_cv.shape)
print("Shape of X_val_processed_cv:", X_val_processed_cv.shape)
print("Shape of X_test_processed_cv:", X_test_processed_cv.shape)
```

### TF-IDF Feature Engineering and Logistic Modelling

In this section, we apply Term Frequency-Inverse Document Frequency (TF-IDF) vectorization to our dataset as part of the feature engineering process. TF-IDF is a popular technique in Natural Language Processing (NLP) that transforms text data into a numerical format by assigning weights to words based on two factors: term frequency (how often a word appears in a document) and inverse document frequency (how unique or rare the word is across all documents). This method allows us to capture the significance of each term within the context of the entire dataset, enhancing the model's ability to identify sentiment-driven patterns.

TF-IDF is particularly valuable for this analysis because our exploratory analysis showed that generic sentiment scoring tools like TextBlob didn’t align well with actual review sentiment labels. By combining positive and negative reviews in a single column and applying TF-IDF vectorization, we can better capture unique language patterns associated with each sentiment type. This refined representation aims to improve model performance by providing contextually relevant and precise features tailored to sentiment prediction.

We will integrate TF-IDF with our logistic regression model and assess its impact on predictive accuracy, comparing the results to our baseline and reduced logistic models.

#### Applying Log Transform, Merging Text Columns and Data Splitting for Model Preparation

```python
# Making a copy 
df_tfidf = df.copy()

# Defining the features to drop based on redundancy and correlation analysis
features_to_drop = [
    'additional_number_of_scoring', 'days_since_review', 'hotel_address_amsterdam', 
    'hotel_address_barcelona', 'hotel_address_london', 'hotel_address_milan', 
    'hotel_address_paris', 'hotel_address_vienna', 'reviewer_nationality_other', 'day'
]

# Dropping the specified redundant features
df_tfidf.drop(columns=features_to_drop, inplace=True)

# Combining positive and negative reviews into 'combined_review' for vectorization
df_tfidf['combined_review'] = df_tfidf['positive_review'].fillna('').astype(str) + ' ' + df_tfidf['negative_review'].fillna('').astype(str)
df_tfidf['combined_review'].fillna('No Review', inplace=True)

# Identifying skewed numerical features for log transformation
skewed_features = [
    'total_number_of_reviews', 
    'review_total_negative_word_counts', 
    'review_total_positive_word_counts', 
    'nights_stayed'
]

# Applying log transformation to skewed features
for feature in skewed_features:
    df_tfidf[feature] = np.log1p(df_tfidf[feature])

# Redefining the feature lists after dropping columns
numerical_features_tfidf = [
    'total_number_of_reviews', 
    'review_total_negative_word_counts', 
    'review_total_positive_word_counts', 
    'nights_stayed', 
    'average_score', 
    'latitude', 
    'longitude', 
    'month', 
    'year'
]

binary_features_tfidf = [
    'leisure_trip', 'reviewer_nationality_australia', 'reviewer_nationality_ireland', 
    'reviewer_nationality_uae', 'reviewer_nationality_uk', 'reviewer_nationality_usa'
]

# Defining target variable and features, excluding positive_review and negative_review at this stage
X_tfidf = df_tfidf.drop(['reviewer_score', 'positive_review', 'negative_review'], axis=1)
y_tfidf = df_tfidf['reviewer_score']

# Train/test split
# First split: 80% of the data for training/validation, 20% for testing
from sklearn.model_selection import train_test_split
X_train_val_tfidf, X_test_tfidf, y_train_val_tfidf, y_test_tfidf = train_test_split(X_tfidf, y_tfidf, test_size=0.2, random_state=12)

# Second split: split the remaining 80% into 60% training and 20% validation
X_train_tfidf, X_val_tfidf, y_train_tfidf, y_val_tfidf = train_test_split(X_train_val_tfidf, y_train_val_tfidf, test_size=0.25, random_state=12)

# Print shapes of the split data to verify the train/validation/test splits
print("Shape of X_train_tfidf:", X_train_tfidf.shape)
print("Shape of X_val_tfidf:", X_val_tfidf.shape)
print("Shape of X_test_tfidf:", X_test_tfidf.shape)
```

#### TF-IDF Vectorization and Scaling

In this step, we apply TF-IDF vectorization. As mentioned above, it will normalize word frequency by adjusting the weight of each term based on its frequency within a document relative to its frequency across the entire corpus. 

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Starting time
start_time = time.time()

# Scaling numerical features
scaler_tfidf = StandardScaler()
X_train_numerical_scaled_tfidf = scaler_tfidf.fit_transform(X_train_tfidf[numerical_features_tfidf])
X_val_numerical_scaled_tfidf = scaler_tfidf.transform(X_val_tfidf[numerical_features_tfidf])
X_test_numerical_scaled_tfidf = scaler_tfidf.transform(X_test_tfidf[numerical_features_tfidf])

# Vectorizing the combined_review text feature with TF-IDF
tfidf_vectorizer = TfidfVectorizer(tokenizer=lambda text: my_tokenizer(text, 'pos') + my_tokenizer(text, 'neg'),
                                   max_features=500, min_df=10, token_pattern=None)
X_train_text_tfidf = tfidf_vectorizer.fit_transform(X_train_tfidf['combined_review'])
X_val_text_tfidf = tfidf_vectorizer.transform(X_val_tfidf['combined_review'])
X_test_text_tfidf = tfidf_vectorizer.transform(X_test_tfidf['combined_review'])

# Dropping 'combined_review' column from the train, validation, and test sets
X_train_tfidf = X_train_tfidf.drop(columns=['combined_review'])
X_val_tfidf = X_val_tfidf.drop(columns=['combined_review'])
X_test_tfidf = X_test_tfidf.drop(columns=['combined_review'])

# Concatenating scaled numerical, binary, and TF-IDF features for each set
X_train_processed_tfidf = np.concatenate((X_train_numerical_scaled_tfidf, X_train_text_tfidf.toarray(), X_train_tfidf[binary_features_tfidf].values), axis=1)
X_val_processed_tfidf = np.concatenate((X_val_numerical_scaled_tfidf, X_val_text_tfidf.toarray(), X_val_tfidf[binary_features_tfidf].values), axis=1)
X_test_processed_tfidf = np.concatenate((X_test_numerical_scaled_tfidf, X_test_text_tfidf.toarray(), X_test_tfidf[binary_features_tfidf].values), axis=1)

# Ending timing
end_time = time.time()

# Printing runtime and shape information
print(f"Processing runtime: {end_time - start_time:.2f} seconds")
print("\nShape of X_train_processed_tfidf:", X_train_processed_tfidf.shape)
print("Shape of X_val_processed_tfidf:", X_val_processed_tfidf.shape)
print("Shape of X_test_processed_tfidf:", X_test_processed_tfidf.shape)
```

Unigrams represent single words, while bigrams and trigrams capture sequences of two and three words, respectively. Including bigrams or trigrams can help capture common phrases or expressions that may better reflect the sentiment or specific content of the reviews. By gathering phrases, n-grams can help the model recognize combinations of words that convey meaning beyond individual terms. For example, the phrase *not good* conveys a negative sentiment that a unigram model might miss if it only sees *not* and *good* separately.

We’ll first use `ngram_range(1,2)`, including unigrams and bigrams, followed by `ngram_range(1,3)`, which also incorporates trigrams, to assess how these ranges impact sentiment prediction accuracy.

### Random Forest Analysis with Optimized Hyperparameters

Random Forest is an ensemble learning method that enhances predictive accuracy and model stability by combining the outputs of multiple decision trees. This technique leverages *bagging*, where each tree is trained on a random subset of data and features. By averaging the predictions from these individual trees, Random Forest reduces the variance typically found in single decision trees, leading to improved generalization and reduced overfitting, especially on high-dimensional or noisy datasets.

In this section, we implement a Random Forest Classifier with optimized hyperparameters to enhance model performance. To streamline training efficiency, we conducted a Randomized Search to identify effective hyperparameter combinations for the Random Forest model. While a comprehensive Grid Search would systematically test all parameter options, Randomized Search provides a time-efficient alternative by evaluating a limited number of random configurations across specified ranges. This approach balances thoroughness with efficiency, aligning with the goals of this experiment.

#### Randomized Parameter Search

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

# Starting time
start_time = time.time()

# Defining the parameter grid for RandomizedSearchCV
param_distributions_rf = {
    'n_estimators': [50, 100],            # Number of trees in the forest
    'max_depth': [5, 10, 15, None],                 # Maximum depth of each tree
    'min_samples_split': [2, 5, 10],                # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],                  # Minimum number of samples required to be at a leaf node
    'max_features': ['sqrt', 'log2']              # Number of features to consider at each split
}

# Instantiating the RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=12)

# Setting up the randomized search with cross-validation
random_search_rf = RandomizedSearchCV(
    estimator=rf_classifier,
    param_distributions=param_distributions_rf,
    n_iter=10,                 # Number of parameter settings sampled (adjust for thoroughness)
    cv=3,                      # 3-fold cross-validation
    random_state=42,
    n_jobs=-1                  # Utilize all cores for faster computation
)

# Fitting RandomizedSearchCV to the training data
random_search_rf.fit(X_train_processed_cv, y_train)

# Outputing the best hyperparameters
print("Best Hyperparameters for Random Forest:", random_search_rf.best_params_)

# Evaluating the best model on the training and test data
train_accuracy_rf = random_search_rf.best_estimator_.score(X_train_processed_cv, y_train)
test_accuracy_rf = random_search_rf.best_estimator_.score(X_test_processed_cv, y_test)

# Ending time
end_time = time.time()

# Printing runtime and accuracy 
print("\nRandom Forest Training Accuracy with Best Hyperparameters:", train_accuracy_rf)
print("Random Forest Test Accuracy with Best Hyperparameters:", test_accuracy_rf)
print("\nRuntime: {:.2f} seconds".format(end_time - start_time))
```

### Feature Analysis from Gradient Boosting Model

Given that the Gradient Boosting model achieved the highest performance among all models, we will perform feature analysis on it to uncover additional insights. Gradient Boosting’s approach of sequentially building trees allows it to capture complex patterns in the data, retaining a broad range of features that contribute to predictive accuracy.

This analysis will enable us to explore feature importance across both text and structured data, potentially revealing new drivers of customer sentiment. By understanding the features emphasized in this high-performing model, Hospitality Insights Group can gain a comprehensive view of the factors that influence guest satisfaction, complementing the findings from previous models.

```python
# Extracting feature importance from the trained Gradient Boosting model
importances = gb_classifier.feature_importances_

# Defining feature names based on the input features
tfidf_feature_names = [f"TFIDF_{i}" for i in range(1, X_train_text_cv.shape[1] + 1)]  # TF-IDF feature names
feature_names = numerical_features + binary_features + tfidf_feature_names  # Combined list of feature names

# Checking if there’s a mismatch between feature names and importances
if len(feature_names) != len(importances):
    # If there is a mismatch, add placeholder names for the extra features
    extra_features_count = len(importances) - len(feature_names)
    extra_feature_names = [f"Extra_Feature_{i+1}" for i in range(extra_features_count)]
    feature_names.extend(extra_feature_names)

# Ensuring that feature names and importances are aligned
if len(feature_names) == len(importances):
    # Create a DataFrame for feature importance
    feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
    feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

    # Displaying the top 20 most important features
    print("Top 20 Feature Importances:")
    print(feature_importance_df.head(20))

    # Plotting the top 20 most important features
    plt.figure(figsize=(10, 8))
    plt.barh(feature_importance_df['Feature'].head(20), feature_importance_df['Importance'].head(20), color='cornflowerblue')
    plt.gca().invert_yaxis()  # Invert y-axis for a descending order
    plt.xlabel("Importance")
    plt.title("Top 20 Feature Importances from Gradient Boosting")
    plt.grid(True, which='both', axis='x', linestyle='--', color='gray', alpha=0.7, zorder=0)
    plt.show()
else:
    print("Feature name count still does not match the number of importances.")
```
